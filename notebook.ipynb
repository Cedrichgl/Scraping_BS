```python
#listes 
Marque = []
Modele = []
Carburant = []
Kilometrage = []
Annee = []
Ville = []
Prix = []


#scraper les pages paginées
for page in range(2, 301):  # change 6 par le nombre de pages que tu veux
    print(f"\n--- Traitement de la page {page} ---")
    url = f"https://www.auto-selection.com/voiture-occasion/annonces-voiture.html?page={page}"
    response = requests.get(url)

    if response.status_code != 200:
        print(f"Erreur lors du chargement de la page {page}")
        continue

    contenu = bs(response.content, 'html.parser')
    annonces = contenu.find_all('div', {'class': 'item clearfix'})

    for annonce in annonces:
        marque = annonce.find('div', {'class': 'title'})
        Marque.append(marque.text.strip() if marque else '')


        modele = annonce.find('div', {'class': 'nobold subtitle'})
        Modele.append(modele.text.strip() if modele else '')

        prix = annonce.find('div', {'class': 'price'})
        Prix.append(prix.text.strip() if prix else '')

        ville = annonce.find('div', {'class': 'localisation'})
        Ville.append(ville.text.strip() if ville else '')

        type_carburant = ''
        nombre_kilometrage = ''
        annee_voiture = ''

        caracteristiques = annonce.find_all('div', {'class': 'caracteristiques list-inline spacer-bottom'})
        for div in caracteristiques:
            txt = div.text.strip()
            if any(carb in txt for carb in ['Essence', 'Diesel', 'Electrique']) and not type_carburant:
                type_carburant = txt
            elif 'km' in txt and not nombre_kilometrage:
                nombre_kilometrage = txt
            elif any(str(yr) in txt for yr in range(1980, 2026)) and not annee_voiture:
                annee_voiture = txt
        Carburant.append(type_carburant)
        Kilometrage.append(nombre_kilometrage)
        Annee.append(annee_voiture)

    time.sleep(2)
```

```python
print('Total Marque : ',len(Marque))
print('Total Modele : ',len(Modele))
print('Total Carburant : ',len(Carburant))
print('Total Prix : ',len(Prix))
print('Total Ville : ',len(Ville))
print('Total Kilometrage: ',len(Kilometrage))
print('Total Annee : ',len(Annee))
```
